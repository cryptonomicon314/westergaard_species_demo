## A.1. Examples

The following examples ilustrate somer of the capabilities of this library.
All analyses are taken from the book
[*Introduction to Tonal Theory*](http://www.amazon.com/Introduction-Tonal-Theory-Peter-Westergaard/dp/0393093425), by Peter Westergaard.

If you run them yourself, using the `ishow()` function instead of the you'll note that the examples take a non-trivial amount of time to typeset (sometimes up to 11 seconds in my machine), especially if you generate SVGs instead of PNGs. This is because in this case we must invoke an externsal program, namely
[ImageMagick](http://www.imagemagick.org/script/index.php) for PNGs (moderately fast) or
[Inkscape](https://inkscape.org/en/) (very slow and annoying for this purpose, as it invokes a GUI that can't be suppressed and temporarily messes your window layout).

When the examples are run with the `ishow_()` function, the typesetting is faster, as the timestamps show.
This is because LilyPond crops the images itself. However, you can only use the `ishow_()` function when the output fits in a single system. In the case of Westergaard's examples this always happens unless you manipulate the size or the width of the score.

Anyway, even if you use the slower `Ã¬show()` function, it is still much faster than creating them by hand in LilyPond or a graphical notation editor, such as
[Sibelius](http://www.avid.com/US/products/family/sibelius),
[Finale](http://www.finalemusic.com/) or
[Musescore](https://musescore.org/).

### Examples 1 and 2

These examples show 2 ways of understanding the same line.
Note that in example 1, one needs 7 stages to understand the line while in example 2 one needs only 5 steps.
This makes the explanation in example 2 more efficient and preferrable.

**Example 1; Page 85, 7th diagram:**

```python
yaml_P85_7 = """
key: c major
lines:
    - name: lower
      clef: bass
      notes: c1 e d g e f g g c

stages:
    - lower: A1|9
    - lower: A2|1
    - lower: A3|7
    - lower: B3|2
    - lower: B1|5; B1|8
    - lower: B2|3; B4|6
    - lower: B3|4
"""
score_P85_7 = lilypondfile_from_yaml(yaml_P85_7)
ishow_(score_P85_7)
```

**Example 2; page 85, 8th diagram:**

```python
yaml_P85_8 = r"""
key: c major
lines:
    - name: lower
      clef: bass
      notes: c1 e d g e f g g c

stages:
    - lower: A1|9
    - lower: A2|1
    - lower: A3|7
    - lower: B4|3,5,6
    - lower: B3|2; B3|4; B1|8
"""
score_P85_8 = lilypondfile_from_yaml(yaml_P85_8, False)
ishow_(score_P85_8)
```

## A.2 Pitches and Accidentals

Abjad supports the same pitch names as the english version of LilyPond. You can find a reference
[here](http://www.lilypond.org/doc/v2.18/Documentation/notation/writing-pitches#note-names-in-other-languages).

```python
staff = Staff("c' d' e' f' g' a' b' c''")
ishow_(staff)
```

Octaves are notated as in LilyPond. Note that Abjad uses notes in absolute instead of relative mode. Absolute mode is also the default in LilyPond, but most users use it in relative mode. A reference can be found
[here](http://www.lilypond.org/doc/v2.18/Documentation/notation/writing-pitches#absolute-octave-entry).

```python
staff = Staff("c1 c'1 c,1")
attach(Clef("bass"), staff)
ishow_(staff)
```

Accidentals are of course also supported, as in the example below. The supported accidentals are those of the English version os LilyPond. Once again, you can find a reference here.
[here](http://www.lilypond.org/doc/v2.18/Documentation/notation/writing-pitches#note-names-in-other-languages).

Since we are mostly interested in *tonal* counterpoint, we will ignore the quarter tone accidentals.

```python
staff = Staff("b'2 bs' bf' b' bff' bss' bqf' bqs' btqf' btqs'")
ishow_(staff)
```

## A.2. Clefs

The following code lists all supported clefs.

```python
display_html(HTML("<p>Supported clefs and middle C (alphabetical order)</p>"))
for name in sorted(Clef._clef_name_to_middle_c_position.keys()):
    staff = Staff("c'1")
    clef = Clef(name=name)
    attach(clef, staff[0])
    display_html(HTML("<ul><li>{}:</li></ul>".format(name)))
    ishow_(staff, pdf=False)
```
